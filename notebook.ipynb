Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# Import necessary packages
import pandas as pd
import numpy as np

# Begin coding here ...
# Use as many cells as you like
```

```python
# importing and load price data

df_price = pd.read_csv('data/airbnb_price.csv')

df_price.info()
```

```python
# import and load room type data

df_room_type = pd.read_excel('data/airbnb_room_type.xlsx') 

df_room_type.info()
```

```python
# import and load reviews data

df_reviews = pd.read_csv('data/airbnb_last_review.tsv', sep='\t') 
df_reviews.info()
```

```python
# merging dataframes

df_airbnb = df_price.merge(df_room_type, on='listing_id') 

df_airbnb.head()
```

```python
# merge 

df_airbnb = df_airbnb.merge(df_reviews, on='listing_id') 

df_airbnb.head()
```

```python
# Convert the last_review column to datetime 

df_airbnb['last_review'] = pd.to_datetime(df_airbnb['last_review']) 

df_airbnb.head()
```

```python
# early and latest review dates 

earliest_review_date = df_airbnb['last_review'].min() 
last_review_date = df_airbnb['last_review'].max()

print(earliest_review_date)
print(last_review_date)
```

```python
# check values for room type    
df_airbnb['room_type'].value_counts()
```

```python
# convert all values to lowercase

df_airbnb['room_type'] = df_airbnb['room_type'].str.lower() 

# count the number of times 'private room' appears in the room type column
private_room_count = df_airbnb['room_type'].value_counts().get('private room', 0)

private_room_count
```

```python
# strip dollars from price
df_airbnb['price'] = df_airbnb['price'].str.replace(' dollars', '')
df_airbnb.head()
```

```python
# cast price column to float

df_airbnb['price'] = df_airbnb['price'].astype('float') 
df_airbnb.info()
```

```python
# Average listing price

average_listing_price = round(df_airbnb['price'].mean(), 2)

average_listing_price
```

```python
# Creating dataframe with solutions

# review dates dataframe

insights = {
    'first_reviewed': [earliest_review_date],
    'last_reviewed': [last_review_date],
    'nb_private_rooms': [private_room_count],
    'avg_price': [average_listing_price]
}

# Create a dictionary with the insights
# insights = {
#     'private_room_count': [private_room_count],
#     'avg_price': [average_listing_price],
#     'total_listings': [total_listings],
#     'latest_review_date': [latest_review_date]
# }

# Create the DataFrame
review_dates = pd.DataFrame(insights)

review_dates
```
